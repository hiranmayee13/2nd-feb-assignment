```python
"""Q2. What is the difference between the functions loc and iloc in pandas.DataFrame?"""
Ans.
    loc: This function is primarily label-based. It is used for selecting rows and columns using labels 
          or boolean arrays. 
        When using loc, you provide the row and column labels explicitly. It includes both the start and 
        end labels in the selection.
        
    iloc: This function is primarily integer-based. It is used for selecting rows and columns using integer 
           positions. 
        When using iloc, you provide the row and column positions explicitly.
        It includes the start position but excludes the end position in the selection.
```

```python
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]})
print(df.loc[1, 'B'])  # Output: 5
```

```python
"""Q3. Reindex the given dataframe using a variable, reindex = [3,0,1,2] and store it in the variable, new_df
then find the output for both new_df.loc[2] and new_df.iloc[2]."""
import pandas as pd
df = pd.DataFrame({'A': [1, 2, 3,4], 'B': [4, 5, 6,3], 'C': [7, 8, 9,6],'D':[1,4,6,9]})
reindex = [3,0,1,2]
new_df=df.reindex(reindex)
new_df
```

```python
new_df.iloc[2]
```

```python
new_df.loc[2]
```

```python
When accessing a row using new_df.loc[2], we retrieve the row with the label/index 2. 
When using new_df.iloc[2], we retrieve the row at the integer position 2
```

```python
"""Q6. What do you understand about the windows function in pandas and list the types of windows
functions?"""

In pandas, the "window functions" refer to a group of operations that are applied to a specific window 
or subset of data within a DataFrame or Series. These functions allow for computation and analysis on a
rolling or expanding basis, taking into account a specified window size or window boundaries.
Rolling Window Functions:

Rolling Mean (rolling().mean())
Rolling Sum (rolling().sum())
Rolling Standard Deviation (rolling().std())
Rolling Min (rolling().min())
Rolling Max (rolling().max())
Rolling Median (rolling().median())
Rolling Apply (rolling().apply())
```

```python
"""Q8. Write a Python program that takes in two dates as input (in the format YYYY-MM-DD) and
        calculates the difference between them in days, hours, and minutes using Pandas time delta. The
        program should prompt the user to enter the dates and display the result."""

DATE_1 = input("Enter Starting Date --> (in the format YYYY-MM-DD) : ")
y1 = int(input("Enter Starting Date Timing in ['Hour']"))
z1 = int(input("Enter Starting Date Timing in ['Minute']"))
DATE_2 = input("Enter Ending Date --> (in the format YYYY-MM-DD) : ")
y2 = int(input("Enter Ending Date Timing in ['Hour']"))
z2 = int(input("Enter Ending Date Timing in ['Minute']"))

date_1 = pd.to_datetime(DATE_1)
date_2 =pd.to_datetime(DATE_2)
time_1 = pd.Timedelta(hours=y1,minutes=z1)
time_2 = pd.Timedelta(hours=y2,minutes=z2)

diff_date = date_2 - date_1
diff_time = time_2 - time_1
print('Difference between these days:',diff_date,diff_time)
```
