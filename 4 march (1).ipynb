```python
"""Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
scatter plot for age and fare columns in the titanic dataset."""
import seaborn as sns
titanic=sns.load_dataset("titanic")
titanic
```

```python
import plotly.express as px

res = px.scatter(titanic, x='age', y='fare')
res.show()
```

```python
"""Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express."""
import seaborn as sns
tips=sns.load_dataset("tips")
tips
```

```python
import plotly.express as px

res = px.box(tips, x='day', y='tip')
res.show()
```

```python
"""Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
column with the color parameter."""
import seaborn as sns
tips=sns.load_dataset("tips")
import plotly.express as px

res = px.histogram(tips, x='sex', y='total_bill',pattern_shape= 'smoker',color='day')
res.show()
```

```python
"""Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
the color parameter."""
import seaborn as sns
iris=sns.load_dataset("iris")
import plotly.express as px

res = px.scatter_matrix(iris,color='species',dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"])
res.show()
```

```python
"""Q5. What is Distplot? Using Plotly express, plot a distplot."""
"""Ans.
    In Plotly Express, the distplot function is used to create a distribution plot, which provides a 
    visual representation of the
    distribution of a single numerical variable.It combines a histogram with a kernel density estimate 
    (KDE) curve."""
from seaborn.distributions import displot
displot(iris,x='sepal_length',y='sepal_width',hue='species')
```
