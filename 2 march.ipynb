```python
"""Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib."""
#Ans 

Matplotlib is a popular Python library used for creating static, animated, and interactive visualizations 
in Python. It provides a wide range of plotting functionality and is extensively used for data exploration, analysis, and presentation.

***Here are five plots that can be plotted using the Pyplot module of Matplotlib:

1.Line Plot: A line plot represents data points connected by straight lines. It is commonly used to 
visualize trends or patterns in data over a continuous interval.

2. Scatter Plot: A scatter plot displays individual data points as markers on a two-dimensional plane. 
It is useful for showing the relationship between two numeric variables and identifying any patterns or clusters.

3. Bar Plot: A bar plot presents categorical data using rectangular bars. It is suitable for comparing 
discrete categories or groups and showing their relative values or frequencies.

4. Histogram: A histogram is used to display the distribution of a continuous variable. It divides the 
range of values into bins and shows the frequency or count of values falling within each bin.

5.Pie Chart: A pie chart represents data as slices of a circular pie, with each slice corresponding to a
different category or proportion. It is helpful for illustrating the composition or relative proportions of 
different parts of a whole.

```

```python
"""Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot."""
#Ans
A scatter plot is a type of plot that displays individual data points as markers on a two-dimensional plane. 
It is commonly used to visualize the relationship between two variables and identify any patterns, trends, or 
clusters in the data.
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y,c='black',s=10)
plt.xlabel("X data")
plt.ylabel("Y data")
plt.title("Scatter plot")
plt.show()
```

```python
"""Q3: Why is the subplot() function used? Draw four line plots using the subplot() function."""
#Ans
The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. 
It allows you to display multiple plots or axes in a single figure, arranged in a grid-like structure.\
The subplot() function takes three arguments: the number of rows, the number of columns, and the index of the current subplot. 
The index starts from 1 and increases row-wise.
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)
plt.plot(x,y)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x,y)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x,y)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.plot(x,y)
```

```python
"""Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot."""
#Ans
A bar plot, also known as a bar chart, is a type of plot that uses rectangular bars to represent categorical 
data. It is commonly used to compare and visualize the values of different categories or groups.

Bar plots are particularly useful for:

1.Comparing categorical data: Bar plots allow easy visual comparison of values across different categories 
or groups. They provide a clear representation of the magnitude or frequency of each category.

2.Showing patterns or trends: Bar plots can reveal patterns, trends, or disparities in data by comparing the 
heights or lengths of the bars. They can help identify which categories have higher or lower values.

3.Presenting discrete data: Bar plots are suitable for discrete or categorical data where the categories are 
distinct and separate.
```

```python
import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.xlabel("companies")
plt.ylabel("profits")
plt.title("bar plot")
```

```python
plt.barh(company,profit)
plt.xlabel("companies")
plt.ylabel("profits")
plt.title("horizontal bar plot")
```

```python
"""Q5: What is a box plot? Why is it used? Using the following data plot a box plot."""
#Ans
A box plot, also known as a box-and-whisker plot, is a type of plot that displays the distribution of a 
dataset using five summary statistics: minimum, first quartile (Q1), median (Q2), third quartile (Q3), 
and maximum. It provides a concise visual representation of the central tendency, spread, and skewness 
of the data.

Box plots are used for various purposes, including:

1.Identifying outliers: Box plots can help detect outliers, which are data points that deviate significantly 
from the rest of the dataset. Outliers are shown as individual points beyond the whiskers of the plot.

2.Comparing distributions: Box plots are useful for comparing the distributions of different datasets or 
different groups within a dataset. They allow easy visual comparison of medians, quartiles, and ranges.

3.Assessing skewness and symmetry: Box plots provide insights into the skewness and symmetry of the data 
distribution. The position and length of the whiskers, as well as the position of the median, can indicate 
the degree of skewness.
```
