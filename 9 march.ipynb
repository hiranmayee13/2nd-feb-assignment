```python
'''Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli
        Distribution and Binomial Distribution?'''

Ans.
    The Bernoulli distribution is a discrete probability distribution that models a single Bernoulli trial, which is an 
    experiment with two possible outcomes: success (usually denoted as 1) or failure (usually denoted as 0). The distribution
    is characterized by a parameter p, which represents the probability of success in a single trial.
    
    The probability mass function (PMF) of the Bernoulli distribution is given by:

            P(X = x) = p^x * (1 - p)^(1-x)

        where X is the random variable representing the outcome (0 or 1), and x is the value that X takes (0 or 1).
        
    An example of the Bernoulli distribution is flipping a fair coin.
    Lets say we define success as getting heads,
    which occurs with a probability of 0.5 (p = 0.5), and failure as getting tails.
    The Bernoulli distribution in this case would have the following probabilities:

            P(X = 0) = (0.5)^0 * (1 - 0.5)^(1-0) = 0.5 (probability of getting tails)
            P(X = 1) = (0.5)^1 * (1 - 0.5)^(1-1) = 0.5 (probability of getting heads)
       The main difference between the Bernoulli distribution and the binomial distribution lies in the number of trials
    considered. The Bernoulli distribution models a single trial with two possible outcomes, whereas the binomial distribution
    models multiple independent and identically distributed (i.i.d.) Bernoulli trials.
    
    The binomial distribution represents the number of successes in a fixed number of independent Bernoulli trials.
    It is characterized by two parameters: n, the number of trials, and p, the probability of success in each trial.
```

```python
'''Q7: Explain uniform Distribution with an example.'''

Ans.
    The uniform distribution is a continuous probability distribution that represents the equal likelihood of observing any
    value within a given interval. In other words, all values within the interval have the same probability density.
    
    The probability density function (PDF) of a uniform distribution is defined as:

            f(x) = 1 / (b - a), for a ≤ x ≤ b
            
    An example of a uniform distribution is rolling a fair six-sided die. In this case, the possible outcomes are the numbers
    1 to 6, and each outcome has an equal probability of 1/6. The uniform distribution can be visualized as a flat line over 
    the interval [1, 6] because each outcome has the same likelihood.
    
    If we want to find the probability of rolling a number between 3 and 5 (inclusive), we can calculate the area under the
    PDF curve within that interval. Since the PDF is a constant value of 1/(6-1) = 1/5 within the interval [3, 5], the 
    probability is equal to the width of the interval multiplied by the height of the PDF:

            P(3 ≤ X ≤ 5) = (5 - 3) * (1/5) = 2/5

    So, in the case of a fair six-sided die, there is a 2/5 probability of rolling a number between 3 and 5.
```

```python
'''Q8: What is the z score? State the importance of the z score.'''

Ans.
    The z-score, also known as the standard score, is a statistical measurement that indicates the number of standard 
    deviations a data point or observation is away from the mean of a distribution. It quantifies how far a value deviates
    from the average in terms of standard deviation units.
    
    
    The z-score, also known as the standard score, is a statistical measurement that indicates the number of standard deviations
    a data point or observation is away from the mean of a distribution. It quantifies how far a value deviates from the average
    in terms of standard deviation units.
    The formula to calculate the z-score for a data point x in a distribution with mean μ and standard deviation σ is:

            z = (x - μ) / σ
            
The importance of the z-score lies in several key aspects:
1.Standardization: 
    The z-score transforms data into a standard normal distribution (mean of 0 and standard deviation of 1), allowing us to
    compare values across different distributions or variables. This standardization facilitates statistical analysis and
    interpretation.  
    
2.Probability calculation:
    The z-score is used to calculate probabilities and determine the percentile or rank of a value within a distribution. 
    By converting a value into a z-score, we can directly look up the corresponding probability or percentile in the standard
    normal distribution table.
    
3.Outlier detection: 
    The z-score is commonly used to identify outliers in a dataset. Observations with z-scores that fall outside a certain
    range (typically beyond ±2 or ±3) are considered unusual or extreme in relation to the mean and standard deviation.
```

```python
'''Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.'''

Ans.
    The Central Limit Theorem (CLT) is a fundamental concept in probability theory and statistics.
    It states that the distribution of the sample means (or sums) of a large number of independent and identically distributed 
    random variables, regardless of the shape of the original population distribution, tends to approximate a normal
    distribution.
    

1.Approximation of population parameters: The Central Limit Theorem allows us to approximate the population mean (μ) using 
the sample mean (X̄) and the population standard deviation (σ) using the sample standard deviation (s). This is crucial for
statistical inference and estimation, as it provides a framework for making inferences about population parameters based on 
sample data.

2.Confidence intervals and hypothesis testing: The Central Limit Theorem forms the basis for constructing confidence intervals
and performing hypothesis tests for population means. The normal distribution approximation allows us to determine the 
likelihood of observing sample means within a specified range and assess the statistical significance of a result.

3.Statistical inference for non-normal populations: The Central Limit Theorem enables us to apply statistical methods
that assume a normal distribution, even when the underlying population distribution is non-normal. This is because, for 
large sample sizes, the sample mean distribution becomes approximately normal, regardless of the population distribution.

4.Stability of statistical estimates: The Central Limit Theorem ensures that the sample mean is a stable and consistent
estimator of the population mean. As the sample size increases, the variability in the sample mean decreases, and it becomes 
more accurate in estimating the true population mean.
```

```python
'''Q10: State the assumptions of the Central Limit Theorem.'''

Ans.
1.Independence:
    The random variables in the sample must be independent of each other. This means that the outcome of one variable should 
    not influence the outcome of another. Independence ensures that each observation provides new information and contributes
    to the overall variability in the sample.
    
2.Identically Distributed: 
    The random variables in the sample must be identically distributed. This means that each variable follows the same
    probability distribution with the same mean and standard deviation. The distribution can be any distribution, as long as
    it has a well-defined mean and finite variance.
    
3.Sample Size: 
    The Central Limit Theorem holds true for large sample sizes. Although there is no strict rule for what constitutes a
    "large" sample size, a commonly accepted guideline is that the sample size should be greater than 30. As the sample size
    increases, the sample mean distribution approaches a normal distribution, regardless of the shape of the original
    population distribution.

4.Finite Mean and Variance:
    The random variables must have a finite mean (μ) and finite variance (σ^2). The mean represents the average value of the 
    variable, while the variance measures the spread or variability of the variable. Finite mean and variance ensure that the 
    distribution of the sample mean approaches a normal distribution as the sample size increases.
```
