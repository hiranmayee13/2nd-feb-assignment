```python
'''Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would
        use each type of test.'''

Ans.
t-test:
    A t-test is used when the sample size is small (typically less than 30) or when the population standard deviation is
    unknown. It is based on the Students t-distribution, which has fatter tails compared to the standard normal
    distribution (z-distribution). The t-test calculates a t-value, which is then compared to the critical values of
    the t-distribution to determine the statistical significance.
    
    e.g :Lets say you want to determine if there is a significant difference in the mean scores of two groups of students
    who received different teaching methods (Group A and Group B). You collect a sample of 20 students from each group
    and measure their test scores. Since the sample sizes are small, you would use a t-test to compare the means of the
    two groups.
    
z-test:
    a z-test is used when the sample size is large (typically greater than 30) and the population standard deviation is
    known. It assumes a normal distribution and utilizes the standard normal distribution (z-distribution) to calculate 
    the z-value for testing the statistical significance.
    
    e.g: Suppose you have data on the heights of adult males in a particular country and want to determine if the mean
    height is significantly different from the average height reported by the national census. You collect a large 
    sample of 500 individuals and know the population standard deviation. In this case, you can employ a z-test to
    compare the sample mean to the known population mean.
```

```python
'''Q2: Differentiate between one-tailed and two-tailed tests.'''

Ans.
One-tailed Test:
    1.	A statistical test in which the alternative hypothesis specifies the direction of the effect 
        (either positive or negative).
    2.The alternative hypothesis is formulated in a specific direction.
    3.Null hypothesis and one-tailed alternative hypothesis.
    4.When there is prior knowledge or a strong theoretical justification for the direction of the effect.
    5.	Testing whether a new drug increases blood pressure.
    
Two-tailed Test:
    1.A statistical test in which the alternative hypothesis does not specify the direction of the effect.
    2.The alternative hypothesis is formulated in both directions.
    3.Null hypothesis and two-tailed alternative hypothesis.
    4.When there is no prior knowledge or theoretical justification for the direction of the effect.
    5.Testing whether a new drug has any effect on blood pressure, regardless of whether it increases or decreases it.
```

```python
'''Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
        each type of error.'''

Ans.
Type 1 error (False Positive):
    A Type 1 error occurs when the null hypothesis is rejected, even though it is true. In other words, its the incorrect
    rejection of a true null hypothesis. It represents a false positive result, indicating that the researcher concludes
    there is a significant effect or difference when, in reality, there isn't.
   * e.g: Suppose a clinical trial is conducted to test the effectiveness of a new drug compared to a placebo. The null
    hypothesis states that there is no difference between the drug and placebo. However, due to random variability or
    other factors, the sample data leads to a rejection of the null hypothesis, concluding that the drug is effective
    when it actually isn't. This would be a Type 1 error.
    
Type 2 error (False Negative):
    A Type 2 error occurs when the null hypothesis is not rejected, even though it is false. Its the failure to reject
    a false null hypothesis. It represents a false negative result, indicating that the researcher fails to detect a 
    significant effect or difference when, in reality, it exists.
  * e.g: Consider a study aiming to determine if a new teaching method improves student performance compared to the 
    traditional method. The null hypothesis states that there is no difference between the two teaching methods. However, 
    due to limited sample size or other limitations, the study fails to find a significant difference and does not reject 
    the null hypothesis, concluding that there is no improvement when, in fact, the new method is effective. This would 
    be a Type 2 error.  
```

```python
'''Q4: Explain Bayes's theorem with an example.'''

Ans.
    Bayess theorem is a fundamental concept in probability theory that describes how to update the probability of an 
event based on new evidence. It involves conditional probabilities and allows us to revise our beliefs or hypotheses in 
light of new information.
    The theorem can be stated as follows:

            ---> P(A|B) = (P(B|A) * P(A)) / P(B)

        Where:
        P(A|B) is the probability of event A given that event B has occurred.
        P(B|A) is the probability of event B given that event A has occurred.
        P(A) and P(B) are the probabilities of events A and B occurring, respectively.
        
e.g:Suppose there is a rare disease that affects 1 in every 10,000 people in a population. You have developed a test to 
    diagnose this disease, which has an accuracy rate of 98%. That is, the test correctly identifies the disease 98% of
    the time and produces a false positive 2% of the time.

Now, imagine a person from this population has tested positive for the disease. What is the probability that the person
actually has the disease?

Let's define the events:

A: The person has the disease.
B: The person tests positive.
We are interested in calculating P(A|B), the probability that the person has the disease given a positive test result.
According to Bayes's theorem:
P(A|B) = (P(B|A) * P(A)) / P(B)

Given the information:
P(A) = 1/10,000 (probability of having the disease)
P(B|A) = 0.98 (probability of testing positive if the person has the disease)
P(B|~A) = 0.02 (probability of testing positive if the person does not have the disease)
P(~A) = 1 - P(A) = 9,999/10,000 (probability of not having the disease)

Let's calculate P(B), the probability of testing positive:
P(B) = P(B|A) * P(A) + P(B|~A) * P(~A)
= 0.98 * (1/10,000) + 0.02 * (9,999/10,000)

Now we can substitute the values into Bayes's theorem:
P(A|B) = (0.98 * (1/10,000)) / [0.98 * (1/10,000) + 0.02 * (9,999/10,000)]
```

```python
'''Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.'''

Ans.
    A confidence interval is a range of values that provides an estimate of the true value of a parameter in statistics.
    It quantifies the uncertainty associated with an estimate by providing a range within which the true value is likely
    to fall, with a certain level of confidence.
    
To calculate a confidence interval, we typically need the following information:

1.Sample data: A representative sample from the population of interest.
2.Point estimate: A single value that estimates the population parameter (e.g., sample mean or proportion).
3.Standard error: A measure of the variability or uncertainty associated with the point estimate.
4.Confidence level: The desired level of confidence for the interval (e.g., 90%, 95%, or 99%).

        Confidence interval = Point estimate Â± (Critical value * Standard error)
    
e.g: Suppose you want to estimate the average height of a population. You collect a random sample of 100 individuals 
    from the population and measure their heights. The sample mean height is 170 cm, and the sample standard deviation
    is 5 cm.

solution :Now, let's say we want to calculate a 95% confidence interval for the population mean height.
          First, we need to determine the critical value associated with a 95% confidence level. For a large sample size
         (e.g., 100), we can use the z-distribution. A 95% confidence level corresponds to a critical value of 1.96 
        (obtained from the z-table).

Next, we calculate the standard error, which is the standard deviation of the sample divided by the square root of the
sample size:
    
Standard error = 5 / â100 = 5/10 = 0.5 cm

Finally, we can calculate the confidence interval using the formula:
Confidence interval = 170 Â± (1.96 * 0.5)
= 170 Â± 0.98
= (169.02, 170.98)

The interpretation of this confidence interval is that we are 95% confident that the true population mean height falls 
within the range of 169.02 cm to 170.98 cm.    
```

```python
from math import sqrt
import scipy.stats as stats


sample_mean = 170
sample_std = 5
n = 100

SEM  = sample_std/sqrt(n)

alpha = 0.05

t_value = stats.t.ppf(1-alpha/2, df=n-1)
print(f't_value for {n} samples with significance level of {alpha} is : {t_value:.2f}')

# Calculate the confidence interval
lower_bound = sample_mean - t_value * SEM
upper_bound = sample_mean + t_value * SEM

# Printing the results
print(f"The 99% confidence interval for population mean is ({lower_bound:.1f}, {upper_bound:.1f})")
```

```python
'''Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error?
        Provide an example of a scenario where a larger sample size would result in a smaller margin of error.'''

Ans.
    The margin of error in a confidence interval represents the range around the sample estimate within which the true 
    population parameter is likely to fall. It quantifies the uncertainty associated with the estimation process.
    A smaller margin of error indicates a more precise estimate.
    
    The effect of sample size on the margin of error, as the sample size increases, the margin of error decreases. 
    This is because a larger sample size provides more information and reduces sampling variability. With more data 
    points, the estimate of the population parameter becomes more precise, resulting in a narrower confidence interval 
    and a smaller margin of error.
    
    e.g: suppose we want to estimate the average height of all students in a university using a random sample of
    students. We take a sample of 50 students and calculate their average height to be 170 cm with a standard deviation
    of 5 cm. We want to calculate a 95% confidence interval for the true population mean height.

   solution: 
Using the formula for the margin of error for a confidence interval, we get:
Margin of error = z * (standard deviation / sqrt(sample size))
For a 95% confidence interval, the critical z-value is 1.96. So, plugging in the values, we get:
Margin of error = 1.96 * (5 / sqrt(50)) = 1.38 cm
This means that we are 95% confident that the true population mean height is within 1.38 cm of our sample mean of 170 cm.
If we had taken a larger sample size of 200 students, the margin of error would have been:
    
Margin of error = 1.96 * (5 / sqrt(200)) = 0.98 cm
```

```python
'''Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average
        of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is
        significantly effective at a 95% confidence level using a t-test.'''

Solution:
    
@Step 1: State the null and alternative hypotheses:

Null hypothesis (H0): The weight loss drug is not significantly effective (mean weight loss = 0).
Alternative hypothesis (Ha): The weight loss drug is significantly effective (mean weight loss â  0).


@Step 2: Set the significance level:

The significance level (Î±) is the threshold that determines how much evidence we require to reject the null hypothesis.
In this case, the significance level is 0.05 or 5% (95% confidence level).

@Step 3: Calculate the test statistic:
The t-test statistic can be calculated using the formula:

                t = (xÌ - Î¼) / (s / ân)

        Where:
        xÌ is the sample mean
        Î¼ is the hypothesized population mean (in this case, 0)
        s is the sample standard deviation
        n is the sample size
        Given:
       xÌ = 6 (sample mean weight loss)
        Î¼ = 0 (null hypothesis mean weight loss)
        s = 2.5 (sample standard deviation)
        n = 50 (sample size)
        Plugging in the values:

        t = (6 - 0) / (2.5 / â50)
        t = 6 / (2.5 / 7.07)
        t = 6 / 0.354
        t â 16.95

@Step 4: Determine the critical value:
Since we are using a t-test, we need to determine the critical value based on the degrees of freedom (df),
which is n - 1. In this case, df = 50 - 1 = 49. Using a t-distribution table or statistical software, we find that the 
critical value for a two-tailed test with a significance level of 0.05 and 49 degrees of freedom is approximately Â±2.009.

@Step 5: Make a decision:
If the absolute value of the test statistic is greater than the critical value, we reject the null hypothesis. 
Otherwise, we fail to reject the null hypothesis.
```

```python
import numpy as np
from scipy.stats import t


alpha = 0.05  # significance level
null_hypothesis = "the drug is not significantly effective"
alternative_hypothesis = "the drug is significantly effective"

mu = 0
sample_mean = 6
sample_std = 2.5
n = 50
df = n - 1  # degrees of freedom

t_score = (sample_mean - mu) / (sample_std / np.sqrt(n))
p_value = 2 * (1 - t.cdf(abs(t_score), df))

print(f"t-statistic: {t_score:.4f}")
print(f"p-value: {p_value}")
if p_value < alpha:
    print(f"Reject the null hypothesis. {alternative_hypothesis}.")
else:
    print(f"Fail to reject the null hypothesis. {null_hypothesis}.")
```

```python
'''Q14. In a study of the effects of caffeine on reaction time, a sample of 30 participants had an average
        reaction time of 0.25 seconds with a standard deviation of 0.05 seconds. Conduct a hypothesis test to
        determine if the caffeine has a significant effect on reaction time at a 90% confidence level using a t-test.'''

Solution:
            The t-test statistic can be calculated using the formula:

                    t = (xÌ - Î¼) / (s / ân)
            where,
            xÌ is the sample mean reaction time
            Î¼ is the hypothesized population mean (in this case, 0)
            s is the sample standard deviation of reaction time
            n is the sample size
            
            
            Sample mean (xÌ) = 0.25 seconds
            Sample standard deviation (s) = 0.05 seconds
            Sample size (n) = 30 participants
```

```python
import math
import scipy.stats as stats


null_hypothesis = "Caffeine has no significant effect on reaction time"
alternative_hypothesis = "Caffeine has a significant effect on reaction time"

sample_mean = 0.25  
sample_std_dev = 0.05 
n = 30 
pop_mean = 0.3  
alpha = 0.1  

t_stat = (sample_mean - pop_mean) / (sample_std_dev / math.sqrt(n))

t_crit = stats.t.ppf(1-alpha/2, n-1)

margin_of_error = t_crit * (sample_std_dev / math.sqrt(n))
lower_ci = sample_mean - margin_of_error
upper_ci = sample_mean + margin_of_error

print(f"t-statistic: {t_stat:.3f}")
print(f"t-critical value: {t_crit:.3f}")
print(f"90% Confidence Interval: ({lower_ci:.3f}, {upper_ci:.3f})")

if abs(t_stat) > t_crit:
    print("Reject the null hypothesis")
    print(alternative_hypothesis)
else:
    print("Fail to reject the null hypothesis")
    print(null_hypothesis)
```
